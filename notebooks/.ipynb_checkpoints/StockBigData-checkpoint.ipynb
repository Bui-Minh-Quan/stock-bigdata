```python
data_path = "hdfs://namenode:9000/user/quan/stocks/*.csv"

# Load all the data into a single data frame
df = spark.read.option("header", True).option("inferSchema", True).csv(data_path)

# parse time as date
df = df.withColumn("time", F.to_date("time", "yyyy-MM-dd"))

# Add a "symbol" column extracted from the file name
df = df.withColumn("symbol", regexp_extract(input_file_name(), r"([^/]+)\.csv$", 1))

df.printSchema()
df.show()
```

```python
symbols = [row.symbol for row in df.select("symbol").distinct().collect()]
print(f"Number of data rows: {df.count()}")
print(f"Number of stock symbols: {len(symbols)}")
symbols
```

```python
w = Window.partitionBy("symbol").orderBy("time")
df = df.withColumn("prev_close", F.lag("close").over(w))
df = df.withColumn("return", (F.col("close") - F.col("prev_close")) / F.col("prev_close"))
df.show()
```

### General Ananlysis

```python
summary = df.groupBy("symbol").agg(
    F.min("time").alias("start_date"),
    F.max("time").alias("end_date"),
    F.count("*").alias("num_days"),
    F.mean("close").alias("avg_close"),
    F.stddev("close").alias("std_close"),
    F.mean("return").alias("avg_return"),
    F.stddev("return").alias("volatility"),
    F.mean("volume").alias("avg_volume")
)

summary.show()
```

### Finding the Correlation between Stock Symbols
For this correlation task, i filter down to 20 stocks that are:
- actively traded (high volume)
- have long, continuous data coverage
- are likely to be representive of the market

Also, as this is a demo, we only keep the data from 2010 to 2020.

```python
from pyspark.sql.functions import col
top20_symbols = (
    summary.filter(summary['start_date'] < "2010-01-01") \
       .filter(summary['end_date'] > "2020-01-01") \
       .orderBy(col("avg_volume").desc()).limit(20).select('symbol')
)

symbol_list = [row.symbol for row in top20_symbols.collect()]
print(symbol_list)
```

```python
pdf_sel_20 = df.filter(col('symbol').isin(symbol_list)).filter(col('time') > '2010-01-01').filter(col('time') < '2020-01-01')

print(f"Number of rows: {pdf_sel_20.count()}")
pdf_sel_20.show(10)
```

```python
pivot_df = pdf_sel_20.groupBy('time').pivot('symbol').avg('return')
pivot_pd = pivot_df.toPandas().set_index('time')
corr_matrix = pivot_pd.corr()
corr_matrix
```

## Machine Learning Models for Stock Price Prediction
In this section, I will develop three machine learning models designed to analyze and forecast stock performance:
- Next-Day Closing Price Prediction Model  predicts the exact closing price for the following trading day.
- Next-Day Price Trend Classification Model  predicts whether the next days closing price will increase (1) or not increase (0).

For demonstration purposes, I will focus on two representative stocks: HCM and SSI.

### Model 1: Next-Day Closing Price Prediction
The idea here is that, we will take the data from the previous 7 days (open, high, low, close, volume) and predict the next day close price (close)

#### Data preparation

```python
pdf_hcm = pdf_sel_20.filter(col('symbol') == 'HCM').orderBy('time')
pdf_hcm = pdf_hcm.na.drop()
pdf_hcm.show(5)
```

```python
from pyspark.ml.feature import StandardScaler, VectorAssembler 

assembler = VectorAssembler(inputCols=['open', 'high', 'low', 'close', 'volume'], outputCol = 'features_raw')
pdf_hcm = assembler.transform(pdf_hcm)
pdf_hcm.show(5)
```

```python
scaler = StandardScaler(
    inputCol = 'features_raw',
    outputCol = 'features_scaled',
    withMean = True,
    withStd = True
) # defines the scaling

scaler_model = scaler.fit(pdf_hcm) # computes mean and std from data
pdf_hcm_scaled = scaler_model.transform(pdf_hcm) # applies scaling to create new column
pdf_hcm_scaled.show(5) # 
```

```python
pdf_hcm_scaled = pdf_hcm_scaled.select('time', 'close', 'features_scaled').toPandas()
pdf_hcm_scaled.head()
```

```python
import numpy as np

window = 7
X, y = [], []

features = np.stack(pdf_hcm_scaled["features_scaled"].to_numpy())
targets = pdf_hcm_scaled["close"].to_numpy()

print(features[0])
print(targets[0])

for i in range(window, len(features)):
    X.append(features[i - window: i].flatten()) # flatten past 7 days * 5 features = 35 features
    y.append(targets[i])

X = np.array(X)
y = np.array(y)

print(X[0])
print(y[0])
```

```python
# Split into training and testing sets

from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("RMSE:", mean_squared_error(y_test, y_pred, squared=False))
print("R2 Score:", r2_score(y_test, y_pred))
```

### To Save memory, I'll warp all the process in a few functions

### Prepare data function

```python
def prepare_supervised_data(df, feature_cols, label_col="close", window=7):
    """
    Convert a time series DataFrame into supervised learning format.
    Each sample uses the past "windon" days of data to predict the next day's label
    
    Parameters:
    - df: DataFrame (sorted in time ascending)
    - feature_cols: list of string (ex: ['open', 'high', 'low', 'close', 'volume'])
    - label_col: a string, the colume name to predict (close)
    - window: interger, number of past days to use for prediction
    
    Returns:
    X: a numpy array. Feature matrix of shape (number samples - window, window * number features)
    y: a numpy array. Target vector of shape (number samples - window, )
    """
    
    assembler = VectorAssembler(
        inputCols = feature_cols,
        outputCol = 'feature_raw'
    )
    
    df = assembler.transform(df) # Create the feature (raw) column
    
    
    # Scale features
    scaler = StandardScaler(
        inputCol = 'feature_raw',
        outputCol = 'features_scaled',
        withMean = True,
        withStd = True
    )
        
    scaler_model = scaler.fit(df)
    
    df = scaler_model.transform(df) # Create the scaled feature column
    
    df = df.select('time', 'features_scaled', label_col).toPandas() # Remove unnecessary columns and convert data frame into pandas
    
    features = np.stack(df['features_scaled'].to_numpy())
    targets = df[label_col].to_numpy()
    
    
    X, y = [], []
    
    for i in range(window, len(features)):
        X.append(features[i - window:i].flatten()) # Flatten past days
        y.append(df[label_col].iloc[i])
    
    
    return np.array(X), np.array(y)
```

### Machine Learning Function

```python
def train_regression_model(X, y, test_size=0.2):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, shuffle=False)
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    metrics = {
        "rmse": mean_squared_error(y_test, y_pred, squared=False),
        "r2": r2_score(y_test, y_pred)
    }
    
    return model, y_pred, y_test, metrics 
```

### Combined Function
For convenience reasons, we combine two above functions into one

```python
def build_stock_model(df, feature_cols=['open', 'high', 'low', 'close', 'volume'], window=7):
    """ 
    Full pipeline: prepare data -> train model -> return model and predictions
    """
    X, y = prepare_supervised_data(df, feature_cols, window=window)
    model, y_pred, y_test, metrics = train_regression_model(X, y)
    
    return {
        "model": model,
        "y_pred": y_pred,
        "y_test": y_test,
        "metrics": metrics
    }
```

### Model 2: Next-Day Price Trend Classification Model
We can reuse the logic from the first model, only slightly change the label into 0 (not increase) or 1 (increase)

```python
def prepare_supervised_logistics_data(df, feature_cols, label_col="return", window=7):
    """
    We just reuse the same logic, slightly change the labels
    """
    
    assembler = VectorAssembler(
        inputCols = feature_cols,
        outputCol = 'feature_raw'
    )
    
    df = assembler.transform(df) # Create the feature (raw) column
    
    
    # Scale features
    scaler = StandardScaler(
        inputCol = 'feature_raw',
        outputCol = 'features_scaled',
        withMean = True,
        withStd = True
    )
        
    scaler_model = scaler.fit(df)
    
    df = scaler_model.transform(df) # Create the scaled feature column
    
    df = df.select('time', 'features_scaled', label_col).toPandas() # Remove unnecessary columns and convert data frame into pandas
    
    df[label_col] = (df[label_col] > 0).astype(int)
    
    features = np.stack(df['features_scaled'].to_numpy())
    
    targets = df[label_col].to_numpy()
    
    
    X, y = [], []
    
    for i in range(window, len(features)):
        X.append(features[i - window:i].flatten()) # Flatten past days
        y.append(df[label_col].iloc[i])
    
    
    return np.array(X), np.array(y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def train_logistics_regression_model(X, y, test_size=0.2):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, shuffle=False)
    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    metrics = {
        "accuracy": accuracy_score(y_test, y_pred),
        "report": classification_report(y_test, y_pred)
    }
    
    return model, y_pred, y_test, metrics 
```

```python
def build_stock_logistics_model(df, feature_cols=['open', 'high', 'low', 'close', 'volume'], window=7):
    """ 
    Full pipeline: prepare data -> train model -> return model and predictions
    """
    X, y = prepare_supervised_logistics_data(df, feature_cols, window=window)
    model, y_pred, y_test, metrics = train_logistics_regression_model(X, y)
    
    return {
        "model": model,
        "y_pred": y_pred,
        "y_test": y_test,
        "metrics": metrics
    }
```

```python
hcm_logistics_results = build_stock_logistics_model(pdf_hcm)
print(hcm_logistics_results['metrics'])
```

```python
hcm_logistics_metrics = hcm_logistics_results['metrics']
print("Logistics Regression Model Accuracy for HCM stock Symbol:", hcm_logistics_metrics['accuracy'])
print("Logistics Regression Model Report for HCM stock Symbol:")
print(hcm_logistics_metrics['report'])

sns.heatmap(confusion_matrix(hcm_logistics_results['y_test'], hcm_logistics_results['y_pred']), annot=True, cmap="Blues", fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix: Stock Trend Prediction (HCM)")
plt.show()
```
